#PROJECT : Fraud Detection System using Neural Networks

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt


df = pd.read_csv("creditcard.csv")


print(df.head())
print()
print(df.info())
print()
print(df.describe())
print()
print("Null values:", df.isnull().sum().max())

fraud_count = df['Class'].value_counts()
no_fraud_pct = round(fraud_count[0] / len(df) * 100, 2)
fraud_pct = round(fraud_count[1] / len(df) * 100, 2)
print(f"No Frauds: {no_fraud_pct}% of the dataset")
print(f"Frauds: {fraud_pct}% of the dataset")

print(f"Total transactions: {len(df)}")
print()
print("Original distribution:")
print(df['Class'].value_counts())


class_counts = df['Class'].value_counts().sort_index()
plt.figure(figsize=(6, 4))
plt.bar(class_counts.index, class_counts.values)
plt.title('Class Distribution Before Resampling')
plt.xlabel('Class (0: Legit, 1: Fraud)')
plt.ylabel('Count')
plt.xticks(class_counts.index, ['Legit (0)', 'Fraud (1)'])
plt.show()


df.drop('Time', axis=1, inplace=True)


scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))


X = df.drop('Class', axis=1)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)


sm = SMOTE(random_state=42)
X_train, y_train = sm.fit_resample(X_train, y_train)


print(f"Total samples after SMOTE: {len(y_train)}")
print(f"Fraud samples(1): {sum(y_train == 1)}")
print(f"Non-fraud samples(0): {sum(y_train== 0)}")


fraud_count_smote = pd.Series(y_train).value_counts()


plt.bar(['No Fraud', 'Fraud'], [fraud_count_smote[0], fraud_count_smote[1]], color=['blue', 'red'])
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.title("Fraud vs. No Fraud Transactions After SMOTE")
plt.show()


def build_model(activation='relu', optimizer='adam'):
    model = Sequential([
        Dense(64, input_dim=X_train.shape[1], activation=activation),
        Dense(32, activation=activation),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy',
                  optimizer=optimizer,
                  metrics=['accuracy'])
    return model


activations = ['relu', 'tanh']
optimizers = ['sgd', 'adam', 'rmsprop']
results = []

for act in activations:
    print(f"\n{'='*40}\nActivation Function: {act.upper()}\n{'='*40}")
    for opt in optimizers:
        print(f"\nOptimizer: {opt.upper()}")
        model = build_model(activation=act, optimizer=opt)
        history = model.fit(X_train, y_train,
                            epochs=10,
                            batch_size=256,
                            validation_split=0.2,
                            verbose=0)
        y_pred = (model.predict(X_test) > 0.5).astype(int)
        report = classification_report(y_test, y_pred, output_dict=True)

        print(classification_report(y_test, y_pred))

        results.append({
            'Activation': act,
            'Optimizer': opt,
            'Accuracy': report['accuracy'],
            'Precision': report['1']['precision'],
            'Recall': report['1']['recall'],
            'F1-Score': report['1']['f1-score']
        })


results_df = pd.DataFrame(results)


metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
plt.figure(figsize=(18, 4))

for i, metric in enumerate(metrics):
    plt.subplot(1, 4, i + 1)
    plt.bar(results_df['Optimizer'] + '-' + results_df['Activation'], results_df[metric], color=['purple', 'green', 'orange', 'red'][i])
    plt.title(metric)
    plt.ylim(0, 1)
    plt.xticks(rotation=45)
    if i == 0:
        plt.ylabel('Score (Fraud Class)')

plt.tight_layout()
plt.show()
```
